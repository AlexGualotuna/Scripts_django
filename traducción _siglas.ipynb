```python
archivo = "Jupyter/traduccion_siglas/siglas.xlsx"
```

```python
import pandas as pd
def normalize_text(value):
    if pd.isna(value):
        return ''
    return (
        str(value)
        .replace('\u00a0', ' ')   # NBSP → espacio normal
        .replace('\r', '')
        .replace('\n', '')
        .replace('\t', '')
        .strip()                 # elimina espacios inicio/fin
    )
```

```python
import pandas as pd

df = pd.read_excel(
    f"/home/odoo/Documentos/DESARROLLO/{archivo}",
    header=0,
    sheet_name="Hoja1"
)

with open('./traduccion_siglas/0. Traduccion.txt', 'w') as f:
    for _, item in df.iterrows():

        asignatura = normalize_text(item['ASIGNATURAS'])
        traduccion = normalize_text(item['TRADUCCION'])

        sql = f"""
UPDATE dara_mallas_subject
SET name_en = '{traduccion}'
WHERE TRIM(name) = '{asignatura}'
    AND company_id = 1;
"""
        f.write(sql)
```
