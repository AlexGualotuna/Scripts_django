```python
texto = """
Semana 1:
Tema: Ética Docente y Vocación Profesional
Subtema 1: Importancia de la ética en la profesión docente (1HR)
    Actividad: Lectura y análisis de casos éticos en la docencia, seguido de una discusión grupal para identificar los valores involucrados y reflexionar sobre su aplicación en situaciones reales.
Subtema 2: Desarrollo de la vocación profesional (1HR)
    Actividad: Elaboración de un diario reflexivo sobre la vocación docente, donde los estudiantes podrán expresar sus motivaciones, expectativas y desafíos en su futura profesión.
Subtema 3: Ética y compromiso en la docencia (1HR)
    Actividad: Debate en grupos pequeños sobre la relación entre ética y compromiso en la docencia, basado en lecturas y casos prácticos.
Semana 2:
Tema: Principios y Valores en la Comunicación Educativa
Subtema 1: Rol de los principios y valores en la comunicación educativa (1HR)
    Actividad: Visualización de videos interactivos sobre la importancia de los principios y valores en la comunicación educativa, seguido de una lluvia de ideas en grupo para identificar ejemplos de su aplicación en el aula.
Subtema 2: Habilidades de comunicación efectiva con enfoque en valores (1HR)
    Actividad: Simulación de situaciones de comunicación educativa, donde los estudiantes practicarán habilidades de comunicación efectiva y persuasiva, enfocadas en la transmisión de valores.
Subtema 3: Estrategias para promover valores a través de la comunicación (1HR)
    Actividad: Diseño de una estrategia de comunicación educativa basada en valores, donde los estudiantes planificarán y presentarán una propuesta de actividades y recursos para promover valores en el aula.
"""
```

```python
# Dividir el texto en líneas
texto = "\n".join([line.strip() for line in texto.split("\n")])
lineas = texto.strip().split('\n')

# Estructura para almacenar la información
estructura = []
semana_actual = {}
tema_actual = {}
subtema_actual = {}

# Procesar cada línea
for linea in lineas:
    if linea.startswith("Semana"):
        # Comenzar una nueva semana
        semana_actual = {"semana": linea, "temas": []}
        estructura.append(semana_actual)
    elif linea.startswith("Tema:"):
        # Comenzar un nuevo tema
        tema_actual = {"tema": linea, "subtemas": []}
        semana_actual["temas"].append(tema_actual)
    elif linea.startswith("Subtema"):
        # Comenzar un nuevo subtema
        subtema_actual = {"subtema": linea, "actividades": []}
        tema_actual["subtemas"].append(subtema_actual)
    elif linea.startswith("Actividad:"):
        # Agregar una actividad al subtema actual
        subtema_actual["actividades"].append(linea.lstrip())

for item in estructura:
    print(item['semana'])
    for item2 in item['temas']:
        print(item2['tema'])
        for item3 in item2['subtemas']:
            print(item3['subtema'])
            for item4 in item3['actividades']:
                print(item4)

# Imprimir la estructura resultante
import pprint
#pprint.pprint(estructura)
```
