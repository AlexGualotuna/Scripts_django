```python
import requests
from openpyxl import load_workbook

# Ruta EXACTA del archivo en tu PC (Windows)
#ruta = r"C:\Users\alexander.gualotuna\Downloads\Mallas web links\mallas.xlsx"
ruta = r"/home/odoo/Documentos/DESARROLLO/Jupyter/mallas web/Mallas web links/mallas.xlsx"
print("📂 Leyendo archivo:", ruta)

wb = load_workbook(ruta, data_only=True)
ws = wb.active

# Extraer hipervínculos reales
urls = []

for row in ws.iter_rows():
    for cell in row:
        if cell.hyperlink:
            urls.append(cell.hyperlink.target)

print(f"🔍 Se encontraron {len(urls)} enlaces\n")

# Verificar respuesta HTTP con contador
for i, url in enumerate(urls, start=1):
    try:
        r = requests.get(url, timeout=600)
        print(f"[{i}/{len(urls)}] ✔ {r.status_code} → {url}")
    except Exception as e:
        print(f"[{i}/{len(urls)}] ❌ ERROR → {url} → {e}")
```

```python
import requests
from openpyxl import load_workbook

#ruta = r"C:\Users\alexander.gualotuna\Downloads\Mallas web links\mallas.xlsx"
ruta = r"/home/odoo/Documentos/DESARROLLO/Jupyter/mallas web/Mallas web links/mallas.xlsx"
print("📂 Leyendo archivo:", ruta)

wb = load_workbook(ruta, data_only=True)
ws = wb.active

urls = []

for row in ws.iter_rows():
    for cell in row:
        if cell.hyperlink:
            urls.append(cell.hyperlink.target)

print(f"🔍 Se encontraron {len(urls)} enlaces\n")

for url in urls:
    try:
        # Reemplazo del dominio
        url_pruebas = url.replace(
            "https://academico.udla.edu.ec",
            "https://pruebasacademico.udla.edu.ec"
        )

        r = requests.get(url_pruebas, timeout=600)
        print(f"✔ {r.status_code} → {url_pruebas}")

    except Exception as e:
        print(f"❌ ERROR → {url_pruebas} → {e}")
```
